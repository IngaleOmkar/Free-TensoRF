```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pe_5k_flower = pd.read_csv('../PSNR_Data/PE_5k_flower.csv')
pe_10k_flower = pd.read_csv('../PSNR_Data/PE_10k_flower.csv')
frpe_10k_flower = pd.read_csv('../PSNR_Data/FRPE_10k_flower.csv')
frpe_5k_flower = pd.read_csv('../PSNR_Data/FRPE_5k_flower.csv')
```

```python
# Plot the line graph for the PSNR values of the 5k iterations 
# The data only has one column of psnr values, ordered sequentially from 0 to 4999
# Label the x-axis as the iteration number and the y-axis as the PSNR value
# title the graph as "PSNR values of 5k iterations for the flower dataset"

plt.figure(figsize=(10,5))
plt.plot(pe_5k_flower)
plt.xlabel('Iteration Number')
plt.ylabel('PSNR Value')
plt.title('PSNR values of 5k iterations for the flower dataset (PE)')
plt.show()
```

```python
# Plot the line graph for the PSNR values of the 10k iterations

plt.figure(figsize=(10,5))
plt.plot(pe_10k_flower)
plt.xlabel('Iteration Number')
plt.ylabel('PSNR Value')
plt.title('PSNR values of 10k iterations for the flower dataset (PE)')
plt.show()
```

```python
# Plot the line graph for the PSNR values of the 10k iterations

plt.figure(figsize=(10,5))
plt.plot(frpe_10k_flower)
plt.xlabel('Iteration Number')
plt.ylabel('PSNR Value')
plt.title('PSNR values of 10k iterations for the flower dataset (FRPE)')
plt.show()
```

```python
# Plot the line graph for the PSNR values of the 5k iterations frpe

plt.figure(figsize=(10,5))
plt.plot(pe_5k_flower)
plt.xlabel('Iteration Number')
plt.ylabel('PSNR Value')
plt.title('PSNR values of 5k iterations for the flower dataset (FRPE)')
plt.show()
```

```python
n_10k_psnr = {'PE': 27.771381304927843, 'FRPE': 28.002779190040492}
n_5k_psnr = {'PE': 26.751347360002597, 'FRPE': 26.83838927689309}

# Two groups of bars should be created, one for the 5k iterations and one for the 10k iterations
# The height of the bars should represent the average PSNR value of the iterations
# the y axis should be zommed in on range 25-28 to show differences in the PSNR values
# This can be done by adding the line plt.ylim(25, 28) after the plt.bar() function

plt.figure(figsize=(10,5))
fig, ax = plt.subplots()
barWidth = 0.3
bar1 = [n_5k_psnr['PE'], n_10k_psnr['PE']]
bar2 = [n_5k_psnr['FRPE'], n_10k_psnr['FRPE']]
r1 = np.arange(len(bar1))
r2 = [x + barWidth for x in r1]
plt.bar(r1, bar1, color='b', width=barWidth, edgecolor='grey', label='PE')
plt.bar(r2, bar2, color='r', width=barWidth, edgecolor='grey', label='FRPE')
plt.ylim(26.5, 28.5)
plt.xlabel('Iterations')
plt.ylabel('Average PSNR')
plt.xticks([r + barWidth for r in range(len(bar1))], ['5k', '10k'])
plt.title('Average PSNR values for PE and FRPE for 5k and 10k iterations')
plt.legend()
plt.show()
```
